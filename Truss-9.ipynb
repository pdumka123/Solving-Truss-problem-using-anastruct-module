```python
from math import *

#lets import the systemElemnts from the anastruct library
from anastruct import *
#instatiate the SystemElement object
ss = SystemElements()



# Tuple of coordinates
A,B,C,D,E,F,G,H,I,J = [0,0],[3,4],[6,4],[6,8],[9, 6],[9,12],[12,8],[12,4],[15,4],[18,0]
#lets add truss members at node using add_truss_element method and passing coordinate loaction as argument to 
#define length of each member
ss.add_truss_element(location=[A, B])
ss.add_truss_element(location=[B, D])
ss.add_truss_element(location=[D, F]) 
ss.add_truss_element(location=[F, G])
ss.add_truss_element(location=[G, I])
ss.add_truss_element(location=[I, J])
ss.add_truss_element(location=[J, H])
ss.add_truss_element(location=[H, E])
ss.add_truss_element(location=[E, C])

ss.add_truss_element(location=[C, A])
ss.add_truss_element(location=[B, C])

ss.add_truss_element(location=[C, D])
ss.add_truss_element(location=[D, E])

ss.add_truss_element(location=[E, F])
ss.add_truss_element(location=[E, G])

ss.add_truss_element(location=[G, H])
ss.add_truss_element(location=[H, I])

#this method plot our assemble structure
ss.show_structure()
```

```python
ss.add_support_roll(node_id=1)
ss.add_support_hinged(node_id=7)
ss.show_structure()
```

```python
for i in range(2,7):
    ss.point_load(node_id=i, Fy =-10)
ss.show_structure()
```

```python
ss.solve()
ss.show_reaction_force()
ss.show_axial_force()
```
