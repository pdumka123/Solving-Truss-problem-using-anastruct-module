```python
# Callin the module
from anastruct import *
# Creating structural element object: ss
ss = SystemElements()

# Creating tuple of nodes
A,B,C,D,E = [0, 3],[3, 3],[6,0],[3,0],[0,0]

# Create Structural Element With Nodals
ss.add_truss_element(location=[A, B])
ss.add_truss_element(location=[B, C])
ss.add_truss_element(location=[C, D])
ss.add_truss_element(location=[D, E])
ss.add_truss_element(location=[E, B])
ss.add_truss_element(location=[B, D])

# Viewing the structure
ss.show_structure()
```

```python
# Define Support Conditions
ss.add_support_hinged(node_id=1)
ss.add_support_hinged(node_id=5)
ss.show_structure()
```

```python
# Add External Forces(load) to the Structure
ss.point_load(node_id=4,Fy=-40)
ss.point_load(node_id=3,Fy=-40)

ss.show_structure()
```

```python
# solving structure
ss.solve()
# Plotting results
ss.show_reaction_force()
ss.show_axial_force()
```

```python
import pandas as pd
df = pd.DataFrame(ss.get_element_results())

pd.set_option('display.float_format', lambda x: '%.2f' % x)
df =df[['id', 'length','Nmax']]
df.to_excel('result.xlsx')
```
